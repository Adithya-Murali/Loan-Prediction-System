```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv("train.csv")
tf=pd.read_csv("test.csv")
```

```python
print(df['Gender'].unique())
print(df['Married'].unique())
print(df['Dependents'].unique())
print(df['Education'].unique())
print(df['Self_Employed'].unique())
```

```python
df=df.drop(labels=["Loan_ID"],axis=1)
tf=tf.drop(labels=["Loan_ID"],axis=1)
```

```python
df['Gender']=df['Gender'].fillna("Male")
tf['Gender']=tf['Gender'].fillna("Male")

df['Married']=df['Married'].fillna("No")
tf['Married']=tf['Married'].fillna("No")

df['Dependents']=df['Dependents'].fillna("0")
tf['Dependents']=tf['Dependents'].fillna("0")

df['Self_Employed']=df['Self_Employed'].fillna("No")
tf['Self_Employed']=tf['Self_Employed'].fillna("No")
```

```python
df['Gender']=df['Gender'].map({'Female':0,'Male':1}).astype(int)
tf['Gender']=tf['Gender'].map({'Female':0,'Male':1}).astype(int)

df['Married']=df['Married'].map({'No':0,'Yes':1}).astype(int)
tf['Married']=tf['Married'].map({'No':0,'Yes':1}).astype(int)

df['Education']=df['Education'].map({'Not Graduate':0,'Graduate':1}).astype(int)
tf['Education']=tf['Education'].map({'Not Graduate':0,'Graduate':1}).astype(int)

df['Self_Employed']=df['Self_Employed'].map({'No':0,'Yes':1}).astype(int)
tf['Self_Employed']=tf['Self_Employed'].map({'No':0,'Yes':1}).astype(int)

df['Dependents']=df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3}).astype(int)
tf['Dependents']=tf['Dependents'].map({'0':0,'1':1,'2':2,'3+':3}).astype(int)
```

```python
df.describe(include='all')
```

```python
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
tf['LoanAmount']=tf['LoanAmount'].fillna(df['LoanAmount'].mean())

df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
tf['Loan_Amount_Term']=tf['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())

df['Credit_History']=df['Credit_History'].fillna("1.0").astype(float)
tf['Credit_History']=tf['Credit_History'].fillna("1.0").astype(float)
```

```python
df['Credit_History']=df['Credit_History'].astype(int)
tf['Credit_History']=tf['Credit_History'].astype(int)

df['Property_Area']=df['Property_Area'].map({'Rural':0,'Semiurban':1,'Urban':2}).astype(int)
tf['Property_Area']=tf['Property_Area'].map({'Rural':0,'Semiurban':1,'Urban':2}).astype(int)

df['Loan_Status']=df['Loan_Status'].map({'N':0,'Y':1}).astype(int)
```

```python
df.info()
```

```python
import seaborn as sns

corrmat = df.corr()
top_corr_features = corrmat.index
fig, ax = plt.subplots(figsize=(10,10)) 
g=sns.heatmap(df[top_corr_features].corr(),annot=True,ax=ax)
```

```python
df_X=df.drop('Loan_Status',1)
df_y=df.Loan_Status
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs')
rfe = RFE(model, 8)
fit = rfe.fit(df_X, df_y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
```

```python
from sklearn.model_selection import train_test_split

X,x,Y,y=train_test_split(df_X,df_y)
y=y.values
```

```python
from sklearn.model_selection import cross_val_score

from sklearn.ensemble.gradient_boosting import GradientBoostingClassifier as gb
model=gb()
model.fit(X,Y)
predicted=model.predict(x)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y,predicted))
```

```python
test_pred = pd.DataFrame(model.predict(tf))
#test_pred = pd.DataFrame(test_pred.idxmax(axis = 1))
test_pred.index.name = 'Loan_ID'
test_pred = test_pred.rename(columns = {0: 'Loan_Status'}).reset_index()
test_pred['Loan_Status']=test_pred['Loan_Status'].map({0:'N',1:'Y'})
test_pred['Loan_ID'] = test['Loan_ID']

test_pred.head()
```
